```python
# read in the cvs
species_path = "../Resources/species.csv"
species_df = pd.read_csv(species_path)
species_df.head()
```

```python
# drop all unnecessary columns
clean_species_df = species_df[['Park Name', 'Category', 'Common Names', 'Abundance', 'Conservation Status']]
clean_species_df
```

```python
# rename columns, remove spaces to make them SQl-friendly
clean_species_df = clean_species_df.rename(columns = {"Park Name": "national_park"}).sort_values('national_park')
clean_species_df = clean_species_df.rename(columns={'Category': 'category', 'Common Names': 'animal_name', \
                                 'Abundance': 'abundance', 'Conservation Status': 'conservation_status'})
```

```python
# Review
print('Describe' + '-'*100)
print(clean_species_df.describe())
print('Abundance counts' + '-'*100)
print(clean_species_df['abundance'].value_counts())
print('Category counts' + '-'*100)
print(clean_species_df['category'].value_counts())
print('Status counts' + '-'*100)
print(clean_species_df['conservation_status'].value_counts())
```

```python
# Wow --- there's an extinct species on the list
extinct = clean_species_df.loc[clean_species_df['conservation_status'] == 'Extinct']
extinct
```
