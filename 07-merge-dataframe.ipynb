```python
import pandas as pd
```

```python
csv_path = "Resources/clean_trails_df.csv"
clean_trails_df = pd.read_csv(csv_path)
clean_trails_df.head()
```

```python
csv_path = "Resources/nps_df.csv"
nps_df = pd.read_csv(csv_path)
nps_df.head()
```

```python
nps_park_diff = set(nps_df.national_park).difference(clean_trails_df.national_park)
print(nps_park_diff)
```

```python
trails_park_diff = set(clean_trails_df.national_park).difference(nps_df.national_park)
print(trails_park_diff)
```

```python
merge_clean_trails = pd.DataFrame.merge(nps_df, clean_trails_df, how='inner')
merge_clean_trails.head()
```

```python
unique_parks = merge_clean_trails['national_park'].unique()
unique_parks
len(unique_parks)
```

```python
merge_clean_trails.info()
```

```python
merge_clean_trails['state'].value_counts()
```

```python
trail_stats = merge_clean_trails.drop(columns=['national_park', 'state', 'acres', 'latitude', 'longitude', 'trail_name', 'trailhead_lat', 'trailhead_long'])
trail_stats.head()
```

```python
trail_stats.to_csv('Output/trail_stats.csv', index=False, header=True)
```
