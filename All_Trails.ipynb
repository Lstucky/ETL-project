```python
cleaning_lat = {
    "{'lat':": ""
}

clean_trails_df = clean_trails_df.replace(cleaning_lat, regex=True)
clean_trails_df['trailhead_lat'] = clean_trails_df['trailhead_lat'].str.strip()

clean_trails_df.head()
```

```python
cleaning_lon = {
    "'lng':": "",
    "}": ""
}
clean_trails_df = clean_trails_df.replace(cleaning_lon, regex=True)
clean_trails_df['trailhead_long'] = clean_trails_df['trailhead_long'].str.strip()

clean_trails_df.head()
```

```python
clean_trails_df = clean_trails_df.rename(columns={"name": "trail_name",
                                                    "area_name": "national_park",
                                                    "state_name": "state",
                                                    "elevation_gain": "elevation",
                                                    "difficulty_rating": "difficulty",
                                                    "route_type": "route_type",
                                                    "avg_rating": "rating",
                                                   })
clean_trails_df.head()
```

```python
clean_trails_df['popularity'] = clean_trails_df['popularity'].round(1)
popularity_stats = clean_trails_df['popularity'].describe()
print(popularity_stats)
clean_trails_df.head()
```
