```python
clean_visits_df = park_visits_df.rename(columns = {"Unit Name": "national_park", "Visitors": "visitors", "Unit Code": "park_id"})
clean_visits_df = clean_visits_df.sort_values('national_park')
clean_visits_df
```

```python
clean_visits_df.count()
```

```python
clean_visits_df['national_park'].unique()
```

```python
visits_replace = {
    "Hawai'i": "Hawaii",
    "Wrangell-St.": "Wrangell - St",
}
clean_visits_df = clean_visits_df.replace(visits_replace, regex=True)
clean_visits_df['national_park'] = clean_visits_df['national_park'].str.strip()
clean_visits_df['national_park'].unique()
```

```python
csv_path = "../Resources/nps_df.csv"
nps_df = pd.read_csv(csv_path)
nps_df.head()
```

```python
nps_replace = {
    "and Preserve": "",
    "The ": "the ",
    "Of ": "of "
}
clean_nps_df = nps_df.replace(nps_replace, regex=True)
clean_nps_df['national_park'] = clean_nps_df['national_park'].str.strip()
clean_nps_df['national_park'].unique()
```

```python
merge_nps_df = pd.DataFrame.merge(clean_nps_df, clean_visits_df, how="left")
merge_nps_df.head()
```
